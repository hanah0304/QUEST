Account 클래스
    -> 은행이름, 예금주, 계좌번호, 잔액
    -> __init__ : 예금주, 초기잔액
    -> 은행이름 : SC은행
    -> 계좌번호 : 3자리-2자리-6자리

```python
import random # 계좌번호 랜덤생성을 위한 패키지
import datetime # 입금, 출근시간을 입력하기 위한 패키지

# account 클래스 생성
class Account:
    def __init__(self,deposit_name, balance=0): # 예금주, 최초잔액 입력
        self.bank_name = "SC은행" # 은행명
        self.deposit_name = deposit_name # 예금주
        self.balance = balance # 초기잔액
        self.deposit_count = 0 # 입금횟수

        # 계좌 번호 생성
        num1 = random.randint(100,999) # 3자리
        num2 = random.randint(10,99) # 2자리
        num3 = random.randint(100000,999999) # 6자리
        self.account_number = str(num1) + '-' + str(num2) + '-' + str(num3) # 계좌번호
        self.deposit_result = [] # 입금시간 기록리스트
        self.withdraw_result = [] # 출금시간 기록리스트
        self.lock_state = False # 계좌 잠금상태 : defalut는 해제된 상태.


# 입금 : 최소 1원 이상
    def deposit(self, amount):
        if self.lock_state == True: # 계좌 잠금 확인. 잠겼을 경우 입금 이용 불가.
            print("계좌 잠김.")
        elif amount >= 1: # 입금액이 1원 이상이여야 입금 가능.
            self.balance += amount # 계좌 잔액에 입금액 합산.
            self.deposit_count += 1 # 입금횟수 추가
            self.deposit_result.append((amount, datetime.datetime.now())) # 입금 기록 입력
            print(f"{amount:,}원 입금. 현재 잔액 : {self.balance:,}") # 입금 내역 출력
            if self.deposit_count % 5 == 0: # 입금횟수가 5번째 단위일 때마다 추가
                self.pay_interest() # 이자 함수로 이동.
        else:
            print("계좌 입금은 1원 이상.") # 계좌 입금액이 1원 미만일 경우

# 이자 : 입금 횟수 5회마다 잔고의 1%
    def pay_interest(self):
        interest = self.balance * 0.01 # 잔고의 1% 계산
        self.balance += interest # 잔액에 이자 합산
        print(f"이자 {interest:,.3f}원. 현재 잔액: {self.balance:,}") # 이자를 소수점 3자리까지 잔액과 함께 쉼표까지 출력.

# 출금 : 계좌 잔고 이상 불가
    def withdraw(self, amount):
        if self.lock_state == True: # 계좌 잠금 확인. 잠겼을 경우 입금 이용 불가.
            print("계좌 잠김.")
        elif self.balance > amount: # 잔액이 출금액보다 많을 경우만 가능.
             self.balance -= amount # 잔액에서 출금액 빼기
             self.withdraw_result.append((amount, datetime.datetime.now())) # 출금 기록
             print(f"{amount:,}원 출금. 현재 잔액 : {self.balance:,}") # 출금 내역 출력
        else:
            print("잔액 부족") # 잔액 부족.


# 계좌 정보 출력
    def display_info(self):
      print(f"은행이름:  {self.bank_name}")
      print(f"예금주: {self.deposit_name}")
      print(f"계좌번호: {self.account_number}")
      print(f"잔고: {self.balance:,}")

# 입금 내역 확인
    def deposit_history(self): # 입금액, 입금시간 출력
        for amount, time in self.deposit_result:
            print(amount, time)

# 출금 내역 확인
    def withdraw_history(self): # 출금액, 출금시간 출력
        for amount, time in self.withdraw_result:
            print(amount, time)

# 입출금 동시 확인
    def all_history(self): # 입출금 동시 확인 (입금내역 + 출금내역)
        all = self.deposit_result + self.withdraw_result
        for amount, time in all:
            print(amount, time)

# 루브릭3을 위한 추가 기능 : 송금??
    def transfer(self, other_acc, amount): # 보내고자 하는 계좌, 이체액 입력.
        if self.lock_state == True: # 계좌 잠금 확인. 잠겼을 경우 이체 이용 불가.
            print("계좌 잠김.")
        elif self.balance >= amount: # 잔액이 이체액보다 많을 경우 이용.
            self.withdraw(amount) # 출금
            other_acc.deposit(amount) # 타 계좌 입금
            print(f"{amount:,}원을 {other_acc.bank_name}의 {other_acc.deposit_name}으로 이체.") # 이체 확인메시지
        else:
            print("잔액 부족")

    def lock(self): # 계좌 잠금
        self.lock_state = True
        print("계좌 잠김.")

    def unlock(self): # 계좌 풀기
        self.lock_state = False
        print("계좌 풀림.")

# 계좌신청서
a = Account("jaehyun", 10000)
b = Account("a0", 20000)
a.transfer(b, 3000) # 이체 예시
a.deposit(10000)
a.display_info() # 요약정보
a.deposit_history()
print(a.bank_name)
```

```python

from random import randrange

class FunnyDice:
    def __init__(self, n=6): # 생성자 구현, parameter 6 (인자를 주지 않으면 6면체 주사위 생성)
        self.n = int(n)
        # 인스턴스 변수
        self.numbers = list(range(1, n+1)) # 주사위의 눈들
        self.index = randrange(0, self.n) # number 리스트의 인덱스
        self.val = self.numbers[self.index] # 주사위의 눈

    def throw(self):  # 랜덤으로 주사위 눈이 나오게 함.
        self.index = 3 # 생략하면 에러 발생
        # self.val = 7  -> 생략하면 행운의 숫자가 랜덤으로 나옴.

    def getval(self): # 주사위 눈 변수 val 반환. -> 주사위 눈의 값을 얻을 수 있음.
        return self.val

    def setval(self, val:int): # 사용자가 주사위 눈을 세팅할 수 있게끔 하는 메서드(범위 밖 숫자 세팅 시 오류 발생)
        if val <= self.n: # if val in self.numbers:
            self.val = val

        else:
            msg = f"주사위에 없는 숫자입니다. 주사위는 1 ~ {self.n}까지 있습니다."
            raise ValueError(msg)


def get_inputs():
    n = int(input("주사위 면의 개수 입력: "))
    return n

def main():
    n = get_inputs() # 주사위 면의 수
    mydice = FunnyDice(n) # 주사위 객체
    mydice.throw() # 메서드로 구현.
    print(f"행운의 숫자는? {mydice.getval()}")

if __name__ == '__main__':
    main()
```
